```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from spotipy.oauth2 import SpotifyOAuth
import pandas as pd
import os

SPOTIPY_CLIENT_ID='5deacdaa767c4e5da17115acb337bf68'
SPOTIPY_CLIENT_SECRET='ef6a984b705d42038aa13feae1792ca0'
redir_uri = 'http://localhost:8888/callback'

auth_manager = SpotifyClientCredentials(SPOTIPY_CLIENT_ID,SPOTIPY_CLIENT_SECRET)

scope = "user-library-read"
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(client_id = SPOTIPY_CLIENT_ID,
                                               client_secret= SPOTIPY_CLIENT_SECRET,
                                               scope=scope,
                                              redirect_uri = redir_uri))
results = sp.current_user_saved_tracks()
```

```python
off_set = 0
this_iteration = sp.current_user_saved_tracks(limit=50,offset=off_set)
results_list = []
while this_iteration['items'] != []:
    
    for track in this_iteration['items']:
        artist_name = track['track']['artists'][0]['name']
        song_name = track['track']['name']
        
        track_res_dict = sp.audio_features(track['track']['id'])[0]
        
        track_res_dict['artist'] = artist_name
        track_res_dict['song_name'] = song_name
        track_res_dict['track_id'] = track['track']['id']
        results_list.append(track_res_dict)
        
    off_set+=50
    this_iteration = sp.current_user_saved_tracks(limit=50,offset=off_set)
```

```python
results_df = pd.DataFrame.from_records(results_list)
results_df.to_csv("MattMalloryEraSavedTracks.csv")
```

```python
def download_youtube(artist,track,root_outdir):
    artist = make_string_safe(artist)
    track = make_string_safe(track)
    
    curdir = os.path.abspath(".")
    
    outdir = os.path.join(root_outdir,f"{artist}___{track}")
    
    if not os.path.exists(outdir):
        os.mkdir(outdir)
        

    cmd = f'youtube-dl "ytsearch:{artist} {track}" --extract-audio --audio-format mp3 --audio-quality 0'
    print(cmd)
    os.chdir(outdir)
    os.system(cmd)
    os.chdir(curdir)\
    
def make_string_safe(my_string):
    for char in ["/","\\","(",")","?"]:
        my_string = my_string.replace(char,"")
    return my_string
```
